# Sentiment Analysis

```python
from wordcloud import WordCloud,STOPWORDS
```

```python
def wordcloud_draw(data, color = 'black'):
    words = ' '.join(data)
    cleaned_word = " ".join([word for word in words.split()
                            if 'http' not in word
                                and not word.startswith('@')
                                and not word.startswith('#')
                                and word != 'RT'
                            ])
    wordcloud = WordCloud(stopwords=STOPWORDS,
                      background_color=color,
                      width=2500,
                      height=2000
                     ).generate(cleaned_word)
    plt.figure(1,figsize=(13, 13))
    plt.imshow(wordcloud)
    plt.axis('off')
    plt.show()
    
# print("Positive words")
# wordcloud_draw(flat_list_1[:20],'white')
```

### Polarity calculation

```python
pol_list = [TextBlob(i).sentiment.polarity for i in df['review_body'].fillna("")]
```

```python
df['polarity'] = pol_list
```

```python
sub_list = [TextBlob(i).sentiment.subjectivity for i in df['review_body'].fillna("")]
```

```python
df['positive_negative']= df['polarity'].values >=0
```

```python
pol_pos = df[df['polarity'] > 0.5]['review'].values
```

```python
pol_pos = df[df['polarity'] > 0.5]['review'].values
flat_list_pos_pol = [item for sublist in pol_pos for item in sublist]
wordcloud_draw(flat_list_pos_pol,'white')
```

# Words from reviews that have high polarity rate (positive)

```python
wordcloud_draw(flat_list_pos_pol,'white')
```

# Words from reviews that have high polarity rate (negative)

```python
pol_neg = df[df['polarity'] < -0.5]['review'].values
flat_list_neg_pol = [item for sublist in pol_neg for item in sublist]
wordcloud_draw(flat_list_neg_pol,'black')
```
